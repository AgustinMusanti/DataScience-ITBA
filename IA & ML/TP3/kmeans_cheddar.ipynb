# K-means sobre el dataset cheddar

Traduccion a Python del flujo del TP3 (originalmente en R) sobre el dataset **cheddar**
(queso cheddar del LaTrobe Valley, Australia: 30 muestras, 4 variables numericas).

**Equivalencias R -> Python:**

| R | Python |
|---|---|
| `set.seed()` | `random_state` |
| `kmeans(base, k)` | `KMeans(n_clusters=k).fit(base)` |
| `km$size` | conteo de `km.labels_` |
| `km$cluster` | `km.labels_` (numera desde 0) |
| `km$centers` | `km.cluster_centers_` |


```python
import pandas as pd
from sklearn.cluster import KMeans

# Carga del dataset
cheddar = pd.read_csv("cheddar.csv")
cheddar.head()
```

## Resumen de la base (equivalente a `summary(cheddar)`)

```python
cheddar.describe()
```

## Agrupamiento k-means

`n_clusters` = cantidad de grupos | `random_state` cumple el rol de `set.seed()`

```python
km = KMeans(n_clusters=4, random_state=42835505, n_init=10).fit(cheddar)
```

## Elementos por grupo (equivalente a `km$size`)

```python
pd.Series(km.labels_).value_counts().sort_index()
```

## Cluster de cada observacion (equivalente a `km$cluster`)

Numera desde 0, no desde 1.

```python
km.labels_
```

## Centroides (equivalente a `km$centers`)

```python
pd.DataFrame(km.cluster_centers_, columns=cheddar.columns)
```
